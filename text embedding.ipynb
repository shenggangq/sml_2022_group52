```python
import random
import json
import numpy as np
from gensim.models.doc2vec import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from gensim.models import Word2Vec
```

```python
train = json.load(open('train.json'))
test = json.load(open('test.json'))
```

# Doc2vec

```python
title = []
abstract_2 = []
for d in train:
    title.append(d['title'])
    abstract_2.append(d['abstract'])
tot = title+abstract_2
document = [TaggedDocument(doc,[i])for i,doc in enumerate(tot)]
model=Doc2Vec(document,vector_size = 650,window = 2,min_count = 1)

```

```python
model.infer_vector(np.asarray(tot[1],dtype = 'str'))
```

```python
text = []
for item in train:
    ab = item['abstract']
    tl = item['title']
    tot = tl+ab
    vec = np.zeros(650)
    for w in tot:
        w_l = []
        w_l.append(str(w))
    vec+= model.infer_vector(w_l) 
    text.append(vec)
text = np.array(text)
```

# Word2vec

```python
title = []
abstr = []
for d in train:
    ab = d['abstract']
    tl = d['title']
    title.append(tl)
    abstr.append(ab)
b = title+abstr
model = Word2Vec(sentences = b,vector_size =100,min_count =1)
w_vec = model.wv
```

```python
text = []
for item in train:
    ab = item['abstract']
    tl = item['title']
    tot = tl+ab
    vec = np.zeros(100)
    for w in tot:
        vec+=w_vec[w]
    vec = vec/len(tot)
    text.append(vec)
text = np.array(text)
```
